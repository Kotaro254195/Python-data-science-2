### stack plots

```python
number1=[1,2,3,4,5,6]
number2=[7,4,3,5,6,7]
number3=[6,8,3,7,2,8]
number4=[3,5,7,9,2,4]
number5=[5,3,2,6,9,10]

# stackplot don't allow u to create label in the background

plt.stackplot(number1,number2,number3,number4,number5,colors=['m','k','y','b','c'])

# plt.legend()
plt.show()
```

```python
number1=[1,2,3,4,5,6]
number2=[7,4,3,5,6,7]
number3=[6,8,3,7,2,8]
number4=[3,5,7,9,2,4]
number5=[5,3,2,6,9,10]

plt.plot([],[],label='number1',color='m',linewidth=10)
plt.plot([],[],label='number2',color='k',linewidth=10)
plt.plot([],[],label='number3',color='y',linewidth=10)
plt.plot([],[],label='number4',color='b',linewidth=10)
plt.plot([],[],label='number5',color='c',linewidth=10)

plt.stackplot(number1,number2,number3,number4,number5,colors=['m','k','y','b','c'])

plt.legend()
plt.show()
```

### loading the plot from files

```python
import csv

x=[]
y=[]

with open('12example.txt','r') as csvfile:
    plots=csv.reader(csvfile,delimiter=',')
    
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
    
plt.plot(x,y,label='loaded from the file',color='b',marker='*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('This is Graph')
plt.legend()
plt.show()
```

```python
import numpy as np
x,y=np.loadtxt('12example.txt',delimiter=',',unpack=True)

plt.plot(x,y,label="Eacy Way to Do",color='b',marker='*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('This is Graph')
plt.legend()
plt.show()
```
